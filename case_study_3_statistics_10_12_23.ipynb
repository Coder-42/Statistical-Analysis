```python
# to calculate quartiles
# it is mandatory that our quartiles and percentile should match each and every time

q1=data['Population in thousands (2017)'].quantile(0.25)
q1
# check it should be equal to twe i.e 25 %
```

```python
q2=data['Population in thousands (2017)'].quantile(0.50)
q2
```

```python
q3=data['Population in thousands (2017)'].quantile(0.75)
q3
```

```python
# IQR range calculation
IQR=q3-q1
IQR
```

```python
uf= q3+(1.5*IQR)
uf
```

```python
lf = q1-(1.5*IQR)
lf
```

```python
# only those records which are outliers will be displayed
data[(data['Population in thousands (2017)']<lf)]| (data['Population in thousands (2017)']>uf)
```
